In this notebook, I seek to predict bike counts by using a linear regression model with:

1. Fourier terms to capture multiple seasonalities
2. ARMA model fitted to the errors to capture short term dependencies
3. Potentially add other regressors to improve fit

```r
#Load data manipulation packages
library(tidyverse)
library(ggplot2)

#Load time series library
library(forecast)
library(tseries)
library(TSA)

#Library to detrend data using ols
library(pracma)
library(SpecsVerification)
```

```r
#Function to adjust plot dimensions
set_plot_dimensions = function(width_choice=12, height_choice=6) {
    
        options(repr.plot.width=width_choice, repr.plot.height=height_choice)
    
        }
```

```r
#Read data in
dat = read.csv('bike_sharing_hourly.csv')
```

```r
#Convert date column from type string to type date
dat[,'dteday'] = as.Date(dat[,'dteday'])
```

### Step 1: Transform count and then split data

```r
#Use a power transformation to stabilize variance
c = 1
lambda1 = BoxCox.lambda(dat$cnt)
transformed_count = BoxCox(dat$cnt, lambda=0) + c #Add constant to remove 0's
```

```r
#Visualize transformed counts
set_plot_dimensions()
plot(transformed_count, type='l')
```

```r
dat$transformed_cnt = transformed_count
```

```r
#Create train and test splits
train = dat[dat[,'dteday'] >= '2011-01-01' & dat[,'dteday'] < '2012-10-31', ]
test = dat[dat[,'dteday'] >= '2012-11-01', ]
```

### Step 2: Create forecasting model 

```r
#Fist convert counts to a time series object
ts_count = ts(data = train$transformed_cnt, frequency = 24)
```

```r
#Plot transformed counts
set_plot_dimensions()
autoplot(ts_count)
```

### Step 2: Use STL iteratively to decompose time series

```r
#Next create an STL object using count ts
#s.window specifies how many observations to consider when estimating seasonal component
#Small values indicate quickly changing seasonality. High indicates the opposite.
seas_smoother = c(23, (24*(330/52)+1), (24*(330)+1) ) #13#24+1#*(365.25/12)#(24)+1
trend_smoother = (24*(330)+1)
#stl_obj = stl(x = ts_count, t.window = trend_smoother, s.window = seas_smoother)
#stl_obj %>% autoplot()
```

```r
#Create msts object to encode multiple seasonalities
ts_freq = c(24, 24*(330/52), 24*(330))
msts_count = msts(data = train$transformed_cnt, seasonal.periods = ts_freq)
```

```r
#Plot it
#pdf('../Bikeshare_data/Pictures/Model_comparisons/stl_decomp.pdf')
set_plot_dimensions(14,10)
msts_count %>% mstl(t.window = trend_smoother, s.window = seas_smoother, 
                    iterate = 5, robust=TRUE) %>% 
autoplot() + xlab('Days')
#dev.off()
```

```r
#Check remainders from iterative decomposition
mstl_decomp = mstl(msts_count, s.window = seas_smoother)
```

```r
#Extract remainder and check assumptions
#pdf('../Bikeshare_data/Pictures/Model_comparisons/stl_resid.pdf')
mstl_decomp_resid = remainder(mstl_decomp)

checkresiduals(mstl_decomp_resid)
#dev.off()
```

These plots indicate that a significant portion of the correlation structure is not captured. 

### Step 2b: Which model to use to predict seasonally adjusted data?

The seasonally adjusted data is defined as:

<center> $\text{Trend}_{t} \hspace{0.1cm} + \text{Residual}_{t}$ <center>

This is what it looks like

```r
sadj_dat = mstl_decomp %>% seasadj() 
sadj_dat %>% checkresiduals()
```

These plots indicate two things. The first is that the upward trend indicates a first differencing would be neccessary to make the data stationary. The second is that using lags of the series and of its residuals could potentially account for some of the autocorrelation structure. 

```r
arima_fit = auto.arima(sadj_dat)
checkresiduals(arima_fit)
```

### Step 3: Predict Test set values

```r
#Forecast using the msts object
n_preds = nrow(test)
n_preds
msts_preds = msts_count %>% stlf(method='arima', h=n_preds, s.window = seas_smoother, 
                                 t.window = trend_smoother, biasadj = TRUE) 
```

```r
summary(msts_preds)
```

### Step 4: Plot predictions and compute accuracy

```r
#Plot predictions and fitted values
#pdf('../Bikeshare_data/Pictures/Model_comparisons/stl_preds_total.pdf')
tot_cnt = msts(dat$transformed_cnt, seasonal.periods = ts_freq)
autoplot(tot_cnt, series = 'Observed') + theme(
    plot.title = element_text(size = 20, face = "bold"),
    legend.title=element_blank(), 
    legend.text=element_text(size=13), 
    axis.text=element_text(size=15),
    axis.title=element_text(size=15,face="bold")) + 
    autolayer(msts_preds$mean, series = 'Test Set') +
    autolayer(fitted(msts_preds), series = 'Train Set') + ylab('Log Counts') + 
    ggtitle('Predicting Log Ride Counts using STLF')
#dev.off()
```

## Checking Predictions

```r
#pdf('../Bikeshare_data/Pictures/Model_comparisons/stl_preds_test_set.pdf')
n1 = nrow(test)#+24
msts_preds$mean %>% autoplot(series='Predicted') + theme(
    plot.title = element_text(size = 20, face = "bold"),
    legend.title=element_blank(), 
    legend.text=element_text(size=13), 
    axis.text=element_text(size=15),
    axis.title=element_text(size=15,face="bold")) +
    autolayer(tail(tot_cnt, n = n1), series = 'Observed') + ylab('Log Counts') + 
    ggtitle('Test set Log Ride count Predictions')
#dev.off()
```

MAPE can't be computed as observed and predicted counts contain 0's. Simple solution is to shift all values by 1 (constant). 

```r
#Train set accuracy
fit_vals = fitted(msts_preds)
accuracy(msts_preds, tot_cnt)
```

```r
#Raw test accuracy
#c = 1
preds = msts_preds$mean #+ c
obsvs = test$transformed_cnt #+ c
```

```r
#Check prediction accuracy
accuracy(preds, obsvs)
```

```r
#Manually computing test set MAPE
MAPE = abs(exp(msts_preds$mean) - test$cnt)/test$cnt
```

```r
#Test set MAPE 
mean(MAPE)
```
