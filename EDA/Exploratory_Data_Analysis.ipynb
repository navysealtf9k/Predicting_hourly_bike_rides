```python
#Import data manip modules
import pandas as pd
import numpy as np

#Visualization libraries
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
```

```python
#Read data in
hourly_dat = pd.read_csv('../Data/bike_sharing_hourly.csv', parse_dates=['dteday'], index_col='dteday')
hourly_dat.index
```

```python
#Change frequency coding from none to hourly
n_obs = len(hourly_dat.index)
hourly_dat.index = pd.date_range('2011-01-01', periods=n_obs, freq='H')
```

```python
#Check normative values of data set
hourly_dat.describe()
```

No indication of any outlying or NA values in the data set

## Data Visualization

```python
#Counts over time
plt.figure(figsize=(14,6))
plt.plot(hourly_dat.cnt)
plt.xlabel('Time in Hours', size=15)
plt.ylabel('Count', size=15)
plt.title('Hourly Count of Bike Rides From 2011 - 2012', size=20)
#plt.savefig('Pictures/visualizing_counts.pdf')
```

Two things stand out in this plot. The first is that the number of rides in 2012 is much higher than in 2011. The second is the data seems to contain a lot of seasonality. The number of rides are lower at the start of the year, increases as we get to the middle and eventually dies down towards the end. We postulate that there are a variety of seasonality patterns in the data:

1. Hourly - i.e. we expect more riders in the morning than at night

2. Daily -  More riders on the weekends relative to weekdays?
3. Monthly - Higher number of riders in the summer relative to the winter

### Checking for Daily, Weekly and Monthly Seasonality

### Daily Seasonality

```python
#Lets take a closer look to understand seasonality better. 
grid1 = sns.FacetGrid(hourly_dat, col='weekday', row = 'workingday')
grid1 = grid1.map(plt.scatter, "hr", 'cnt', edgecolor="w")

#plt.savefig('Pictures/weekly_seasonality.pdf')
```

The diagrams above highlight interesting hourly and daily seasonal trends. The first thing to note is we have created an intersection between two variables, weekdays and working days. We do so as working days alone does not tell us if a particular day was a weekend or not. We can infer this by looking at the interactions between the 2 variables mentioned above. It is clear from the picture above that  week days 0 and 6 are the weekend. 

The diagrams above show us that the distribution of rides is unimodal on the weekends (achieving a steady pic midday and dying down slowly towards the end of the day). The week days tell a different story. We notice that there are 2 peaks on these days. They occur from 8-9 am and again between 4 and 5 pm. This is expected as a large proportion of users commute to work during these times of the day. 

```python
#Read daily data in
daily_dat = pd.read_csv('bike_sharing_daily.csv', parse_dates=['dteday'])#, index_col='dteday')
daily_dat.head()
```

```python
#Change frequency coding from none to hourly
n_days = len(daily_dat.index)
daily_dat.index = pd.date_range('2011-01-01', periods=n_days, freq='D')
```

```python
#Lets plot daily ride counts
f, axes = plt.subplots(3, 1, figsize=(13,6), sharex=False)

#Total count plot over time
sns.lineplot(x='dteday', y='cnt', data=daily_dat, ax=axes[0])

#Count plot in 2011
sns.lineplot(x='dteday', y='cnt', data=daily_dat.iloc[1:365,], ax=axes[1])

#Count plot for 3 months in 2011 to identify daily seasonality
sns.lineplot(x='dteday', y='cnt', data=daily_dat.iloc[1:90,], ax=axes[2])
```

The above plot makes it easier to visualize a few things about the data set. The first, is that there is a somewhat curvilinear trend in the data. The second is that multiple seasonalies are involved. It is not obvious that counts exhibit daily seasonality. We saw above that the data does exhibit hourly seasonality. To check if monthly seasonality is present, lets plot resampled (converting daily to monthly) counts.

### Weekly Seasonality

```python
plt.figure(figsize=(14,6))
plt.plot(daily_dat.cnt.resample('W').sum())
plt.xlabel('Time in Weeks', size=15)
plt.ylabel('Count', size=15)
plt.title('Weekly count of bike rides', size=20)
#plt.savefig('Pictures/weekly_counts.pdf')
```

The graph above does not show any obvious cyclical patterns. Lets check if it exhibits monthly trends

### Monthly Seasonality

```python
plt.figure(figsize=(14,6))
plt.plot(hourly_dat.cnt.resample('M').sum())
plt.xlabel('Time in Months', size=15)
plt.ylabel('Count', size=15)
plt.title('Monthly count of bike rides', size=20)

#plt.savefig('Pictures/monthly_counts.pdf')
```

The following graph highlights the monthly seasonality of the count of bike rides in Washington. We notice that, there is a substantial level difference between values in 2011 and 2012. Furthermore, its clear that bike rides are highest in the spring/summer (May - July) and lowest in the winter (November to February). The number of bike rides in 2012 increase between September and October, whilst they appear to be decreasing between September and October of the previous year. 

```python
#Lets look at total ride count by month
plt.figure(figsize=(13,6))
sns.barplot(x='mnth', y='cnt', data=hourly_dat, label='Rides in a month')
plt.xlabel('Month', size=15)
plt.ylabel('Count', size=15)
plt.title('Total Counts by Month', size=20)
#plt.savefig('Pictures/total_monthly_counts.pdf')
```

```python
#Look at ride counts across each month
plt.figure(figsize=(13,6))
sns.boxplot(x='hr', y='cnt', hue='weathersit', data=hourly_dat)
plt.xlabel('Hour of the Day', size=15)
plt.ylabel('Count', size=15)
plt.title('Breakdown of Hourly Ride Counts by Weather', size=20)
#plt.savefig('Pictures/hourly_count_weather_breakdown.pdf')
```

The graph above highlights a few interesting trends. We observe that ride counts are highest during months between June and October. Furthermore, rides are highest in when the weather is good (weathersit 1 and 2) and lowest when it is bad(weathersit 3). Apart from a few riders in the month of January,nobody rides bikes when weather is bad (weathersit 4)

## Type of seasonal models to use?

From the plots above we observe:
1. Hourly seasonality - Peaks at the same time of roughly the same frequency across weekdays 
2. Monthly seasonality - Peaks in the summer and lows in the winter of roughly the same magnitude

This suggests that an appropriate model would have **multiple seasonalities**. Further, each of these seasonalies could be described well with an **additive model**.

### Relationship between Temperature, Wind, Humidity and count

```python
#What type of relationship can be observed between rides and each of these environmental variables?
f, axes = plt.subplots(1, 3, figsize=(14,6), sharex=False, sharey=False)

#Temp vs count
sns.scatterplot(x='atemp', y='cnt', data=hourly_dat, ax=axes[0])
plt.xlabel('Normalized Real Feel Temperature')
plt.ylabel('Count')

#Windspeed vs count
sns.scatterplot(x='windspeed', y='cnt', data=hourly_dat, ax=axes[1])
plt.xlabel('Normalized Windspeed')
plt.ylabel('Count')

#Humidity vs count
sns.scatterplot(x='hum', y='cnt', data=hourly_dat, ax=axes[2])
plt.xlabel('Normalized Humidity')
plt.ylabel('Count')
```

The first plot suggests that temperature and ride count are positviely correlated. As temperatures go up (days are warmer) the number of riders increase. Ride counts fall when temperatures are very high (> 40 degrees celcius). 

The second plot suggests no correlation between windspeed and ride count. It seems that the number of riders are constant, regardles off the windspeed. Like with temperature, ride counts fall as windspeeds approach their maximum value (67mph?). 

The third plot suggests that humidity and ride counts are negatively correlated. This is intuitively plausible. The more moist it is outside, the more sweaty a person is likely to feel. Thus they are more likely to avoid riding bikes in this weather condition. 

## Understanding correlation structure between  variables

```python
#Correlation matrix
sns.set(font_scale=1.3)
cols = ['windspeed', 'atemp','hum', 'cnt']
corr_df = hourly_dat.loc[:,cols].corr(method='pearson')
# corr_df = hourly_dat.corr(method='spearman')

f, ax = plt.subplots(figsize=(14, 10))
sns.heatmap(corr_df, annot=True, linewidths=.5, ax=ax, cmap = "Oranges")
plt.title('Correlations between Counts and Environmental Variables', size=20)
#plt.savefig('Pictures/corr_heatmap.pdf')
```
