In this notebook, I study the statistical properties of the various time series in my data set. My main goals are to:

1. Identify the dominant frequencies of each series. 

2. Fit suitable time series models to each series.

```r
#Load data manipulation packages
library(tidyverse)
library(ggplot2)

#Load time series library
library(forecast)
library(tseries)
library(TSA)

#Library to detrend data using ols
library(pracma)
library(SpecsVerification)
```

```r
#Read data in
dat = read.csv('bike_sharing_hourly.csv')
```

```r
#Convert date column from type string to type date
dat[,'dteday'] = as.Date(dat[,'dteday'])
```

```r
#Create train and test splits
train = dat[dat[,'dteday'] >= '2011-01-01' & dat[,'dteday'] < '2012-10-31', ]
test = dat[dat[,'dteday'] >= '2012-11-01', ]
```

```r
#Check train and test splits
tail(train)
```

```r
#Function to adjust plot dimensions
set_plot_dimensions = function(width_choice=12, height_choice=6) {
    
        options(repr.plot.width=width_choice, repr.plot.height=height_choice)
    
        }
```

```r
#Plot daily counts
set_plot_dimensions()
ggplot(aes(y=cnt, x=dteday), data=train) + geom_line()
#plot(y=train$cnt, x=train$dteday, type = 'l')
```

Bike counts look non-stationary. We observe an upward trend. Bike rides seem to be going up in 2012, compared to 2011. This time series exhibits clear seasonal patterns. Counts go up in the summer and down in the winter. Furthermore, the size of seasonal fluctuations vary with time. Hence a multiplicative model maybe more appropriate here. Lets use a log transform to convert it into a process that can be modelled additively.

```r
#Plot of log counts
set_plot_dimensions()
ggplot(aes(y=log(cnt), x=dteday), data=train) + geom_line()
```

```r
#Transformation
log_tr_cnt = log(train$cnt)
```

Due to the high frequency of the data, it is difficult to specify the periodicity of the data. To do so, we use a periodogram. We identify the largest spikes in it. The period is the inverse of these frequencies.

### Periodicity

```r
pg = periodogram(log_tr_cnt)
```

```r
#Identify where the largest spikes occur
pg_df = data.frame(freq=pg$freq, spec=pg$spec)
pg_df$freq = 1/pg_df$freq #Convert frequencies into the right form
pg_df = round(pg_df[order(-pg$spec),])
head(pg_df, 4)
```

```r
#Function to determine frequency

#Inputs: ts - Vector object to be converted into a time series
get_ts_freq = function(ts) {
    
    pg = periodogram(ts)
    pg_df = data.frame(freq=pg$freq, spec=pg$spec)
    pg_df$freq = 1/pg_df$freq #Convert frequencies into the right form
    pg_df = round(pg_df[order(-pg$spec),])
    
    #Return largest frequency
    freq = pg_df[1,1]
    freq1 = head(pg_df,4)
    print(freq1)
    return(freq)
}
```

The results suggest that an annual frequency (~337) is appropriate if we analyz daily data. If its hourly, then the optimal would be dail frequency (~24 hrs). We use the ndiffs function, which calls on the kpss test sequentially to determine the number of differences to compute to make our time series trend stationary. 

```r
#Use ndiffs function to determine how many differences to compute
freq = pg_df[1,1]
ts_cnt = ts(log_tr_cnt, frequency = freq)
ndiffs(ts_cnt)
```

```r
#Compute and plot seasonal differences
diff_cnt = diff(log_tr_cnt, differences = 1) 

#Plot first difference
diff_cnt %>% ggtsdisplay()
```

The first order difference seems to have eliminated the trend. However we notice that our differenced counts are seasonally non stationary. This is verified by the sinosidual patterns observed in the ACF and PACF plots. Whilst the trend seems more linear, the data still indicates the presence of seasonal variation. It is no longer appropriate to use the ADF and KPSS tests to test for stationarity. This is because these tests check if a time series has a unit root in it. They are not however designed to identify non stationarities induced by seasonality. 

Lets use the nsdiffs function to determine the number of seasonal differences to compute. This may not be appropriate when studying daily data. The periodogram suggests an optimal frequency of 337. Thus a seasonal first difference would require more than 674 observations (we don't have that). I adjust it to 300 to make differencing possible. 

It may also not be appropriate to compare sales on the 1st of January last yeart to sales on the first of January this year. 

```r
#Number of seasonal differences?
ts_cnt = ts(log_tr_cnt, frequency = freq)
nsdiffs(ts_cnt)
```

The test above suggests a seasonal difference of 1. 

```r
seasonal_diff_cnt = diff(ts_cnt, lag = freq)
seasonal_diff_cnt %>% ggtsdisplay()
```

```r
#Check if the process is trend stationary?
seasonal_diff_cnt %>% nsdiffs()
```

The process above looks trend stationary. However it does not look trend stationary. One reason for this is the high frequency of our data. From our EDA conducted previously, we observed that there are distinct weekly and yearly patterns. These cannot be removed by simply computing another difference of the same order.

```r
#Lets look for dominant frequencies in the differenced time series
get_ts_freq(diff_cnt)
```

```r
#Lets perform another round of differencing
seasonal_diff_cnt %>% diff(lag = 12) %>% ggtsdisplay()
```

### 1. Temperature

The plot above is clearly neither trend nor seasonally stationary. Lets first examine its periodicity.

```r
get_ts_freq(train$atemp)
```

The first two values are nonsensical. Thus we use the 3rd largest value of 24 observations a cycle. Next, lets look at transformations we can use to make temperature stationary.

```r
#How many differences to make this proccess seasonally and trend non-stationary?
temp_ts = ts(train$atemp, frequency = 24)
paste0('Number of regular diffs to achieve trend stationarity is: ',ndiffs(temp_ts))
paste0('Number of regular diffs to achieve seasonal stationarity is: ',nsdiffs(temp_ts))
```

```r
#Function to calculate number differences needed to make ts stationary

#Inputs:  ts - Vector to convert into a time series object
#       freq - Integer specifying frequency of time series
make_ts_stationary = function(ts,freq) {
    
    ts_1 = ts(ts, frequency = freq)
    trend_diffs = ndiffs(ts_1)
    seas_diffs = nsdiffs(ts_1)
    
    #Print diffs taken
    print(paste0('Number of regular diffs to achieve trend stationarity is: ', trend_diffs))
    print(paste0('Number of regular diffs to achieve seasonal stationarity is: ',seas_diffs))
    
    #Compute diffs and graph output
    if (seas_diffs == 0) {
        seas_tr_diff = ts_1 %>% diff(lag=1, differences = trend_diffs)
    } 
    else if (trend_diffs == 0) {
        seas_tr_diff = ts_1 %>% diff(lag = freq, differences = seas_diffs) 
    }
    else if(trend_diffs == 0 & seas_diffs == 0) {
        seas_tr_diff = ts_1
    }
    else {
        seas_tr_diff = ts_1  %>% diff(lag = freq, differences = seas_diffs) %>% 
        diff(lag=1, differences = trend_diffs)
    }    
    
    
    #ACF and PACF plots
    seas_tr_diff %>% ggtsdisplay()
}
```

```r
make_ts_stationary(train$atemp, 24)
```

The process now looks trend stationary. However it appears that we haven't captured it all, based on the ACF and PACF plots. The ACF plot contain significant lags at multiples of 24. The same pattern is observed in the PACF plot as well. This suggests that temperature is potentially mutliseasonal as well.

# <center> Multiple Seasonality Adjustment <center>

```r
#Add a quadratic trend to the data
t_sq = seq(1,length(train))

#Add to dataframe
train$trend = t_sq
```

```r
#Estimate trend
fit = lm(cnt ~ trend + I(trend^2) + I(trend^3), data = train)
summary(fit)
```

```r
#Remove trend from count data
detrended_cnt = train$cnt - fit$coefficients[2]*t_sq - fit$coefficients[3]*t_sq^2 - 
fit$coefficients[2]*t_sq^3
```

```r
detrended_cnt %>% ggtsdisplay()
```
